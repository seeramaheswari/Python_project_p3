**Calculating yearly sales trend using group by function**

```python
yearly_sales=dataset.groupby(dataset['Order Date'].dt.year)['Sales'].sum()
print(yearly_sales)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(yearly_sales.index, yearly_sales.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Yearly Sales Trend')
plt.show()
```

**Let's Check For The Repeated Customers**

```python
dataset['Customer ID'].value_counts().sort_values(ascending=False).head(5)
```

```python
Customer_order_count=dataset.groupby('Customer ID')['Order ID'].nunique().sort_values(ascending=False)
Customer_order_count.head(6)
```

```python
Repeated_customers=Customer_order_count[Customer_order_count>1].index
Repeated_customers
```

```python
Repeat_customer_data=dataset[dataset['Customer ID'].isin(Repeated_customers)]
Repeat_customer_data
```
